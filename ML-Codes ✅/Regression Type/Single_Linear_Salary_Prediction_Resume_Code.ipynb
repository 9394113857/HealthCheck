1.	**Before .values:**

2.	**After .values:**

  •	The Pandas DataFrame is converted to a Numpy array (ndarray).
  
  •	Column names and index are lost, and only the numerical values are retained.

3.	•**Summary:**

  •	Before .values: Data is a Pandas DataFrame.
  
  •	After .values: Data becomes a Numpy ndarray, and column names and index are discarded.

4.**Recommended Approach:**

  •	It is generally best to use .values (or .to_numpy()) when preparing data for machine learning models, as Numpy arrays are typically more efficient for processing.

  • If column names and index are important for your task, you should keep the data as a Pandas DataFrame until you're ready for model training.

# User-Prediction without Saved Model:-

```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
from rich.console import Console

# Sample data for training the model (experience vs salary)
X_train = np.array([[1], [2], [3], [4], [5]])  # Experience in years
y_train = np.array([40000, 45000, 50000, 55000, 60000])  # Corresponding salaries

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Create a Console instance for rich text printing
console = Console()

# Print the title to explain the process
console.print("[bold blue] Prediction Model (Before Saving and Pickling)[/bold blue]")
console.print("[italic] This is a test run of the model before saving (pickling) it for future use. [/italic]")

# Print a separator
console.print("===================================================")

# Placeholder for the default experience value
default_experience_value = 1.1  # Changed the default value to 1.1 years

# Prompt the user with a default value and take the input
user_input = input(f"Enter your experience (in years) [default: {default_experience_value}]: ")

# If the user doesn't enter anything, use the placeholder value
if not user_input:
    Experience_Value = default_experience_value
else:
    # Otherwise, use the entered value, converting to float
    Experience_Value = float(user_input)

# Print the entered experience value in green
console.print(f"The entered experience value: [bold green]{Experience_Value}[/bold green]")

# Use the trained model to predict salary based on the user's input experience value
User_Prediction = model.predict([[Experience_Value]])

# Print the predicted salary from the model in bold green
console.print(f"The predicted salary from the model: [bold green]{User_Prediction[0]}[/bold green]")

# Print a separator
console.print("===================================================")

# Print the final predicted salary in bold green
console.print(f"So, the Salary should be: [bold green]{User_Prediction[0]}[/bold green]")  # Placeholder for final output
```

```python
import os
import pickle

# Check if the pickle file already exists
filename = 'single_linear_regression_salary_prediction.pkl'
if os.path.exists(filename):
    # Ask the user for confirmation before saving
    user_input = input(f"File '{filename}' already exists. Do you want to overwrite it? (y/n): ")
    if user_input.lower() == 'y':
        # Save the trained Linear Regression model using Pickle
        pickle.dump(model.fit(x, y), open(filename, 'wb'))
        print(f"Model saved as '{filename}'")
    else:
        print("Model not saved. Skipping.")
else:
    # Save the trained Linear Regression model using Pickle
    pickle.dump(model.fit(x, y), open(filename, 'wb'))
    print(f"Model saved as '{filename}'")
```

3.	pickle.dump() is used to serialize the trained model, converting it into a byte stream, and then writes it to the file specified.
4.	The file is opened in binary write mode ('wb') to ensure that it can properly handle the binary data that represents the trained model.
5.	After execution, the file 'single_linear_regression_salary_prediction.pkl' will contain the serialized representation of the trained Linear Regression model, making it ready for later use, such as making predictions on new data without needing to retrain the model from scratch.

```python
import pickle
import os
from rich.console import Console
from rich.text import Text
from datetime import datetime
import pytz

# Initialize a Console object for rich output
console = Console()

# Function to get current time in IST and formatted date/time string
def get_current_time_ist():
    ist = pytz.timezone('Asia/Kolkata')  # Set IST timezone
    now_ist = datetime.now(ist)  # Get the current time in IST
    formatted_date_time = now_ist.strftime('%Y-%m-%d %H:%M:%S')  # Format the date and time
    return formatted_date_time

# Function to list all model files in the current directory
def list_model_files():
    files = [f for f in os.listdir() if f.endswith('.pkl')]
    return files

# List all the model files
model_files = list_model_files()

# If no model file is found, exit
if not model_files:
    console.print("[bold red]Error: No model files found in the current directory![/bold red]")
    exit()

# If only one model file is found, use it automatically
if len(model_files) == 1:
    model_file = model_files[0]
    console.print(f"[bold cyan]Using default model file: {model_file}[/bold cyan]")

# If multiple model files are found, prompt the user to select one
else:
    console.print("\n[bold yellow]Multiple model files found:[/bold yellow]")
    for i, model in enumerate(model_files, start=1):
        console.print(f"{i}. {model}")

    # Ask the user to select a model
    try:
        choice = int(input(f"\nEnter the number corresponding to the model you want to use (1-{len(model_files)}): "))
        if 1 <= choice <= len(model_files):
            model_file = model_files[choice - 1]
            console.print(f"[bold cyan]Using selected model file: {model_file}[/bold cyan]")
        else:
            console.print("[bold red]Invalid choice. Exiting.[/bold red]")
            exit()
    except ValueError:
        console.print("[bold red]Invalid input. Exiting.[/bold red]")
        exit()

# Load the model
try:
    saved_model = pickle.load(open(model_file, 'rb'))
except Exception as e:
    console.print(f"[bold red]Error loading the model: {e}[/bold red]")
    exit()

# Print the section separator
console.print("\n" + "=" * 60)
console.print("[bold cyan]Model Loaded Successfully![/bold cyan]")
console.print("=" * 60)

# Get current time for displaying
current_time = get_current_time_ist()

# Direct and simple experience input prompt
console.print("\n[bold magenta on white] Enter Experience (in years): [/bold magenta on white]")
Experience_Value = float(input().strip())  # Get user input directly

# Print entered experience value
console.print(f"\n[bold cyan]Entered Experience Value: [/bold cyan][italic yellow]{Experience_Value}[/italic yellow]")
console.print("=" * 60)

# Display current time of execution in the output
console.print(f"[bold green]Execution Time: {current_time}[/bold green]")
console.print("=" * 60)

# Use the model to predict salary based on experience
User_Prediction = saved_model.predict([[Experience_Value]])

# Extract the scalar value from the prediction (it will be a single element array)
predicted_salary = User_Prediction.item()  # Converts the NumPy scalar to a Python scalar

# Print the predicted value from the model
console.print(f"[bold blue]Predicted Salary: [/bold blue][bold green]{predicted_salary:.2f}[/bold green]")
console.print("=" * 60)
```

```python
# Step 6: Extract file metadata and display as a table with Last Modified time converted to IST
# and formatted as dd/mm/yyyy hh:mm:ss AM/PM IST

import pandas as pd
import pytz  # For timezone conversions

# Helper function to convert bytes to human-readable format
def human_readable_size(size):
    for unit in ['B', 'KB', 'MB', 'GB']:
        if size < 1024:
            return f"{size:.1f} {unit}"
        size /= 1024
    return f"{size:.1f} TB"

# Define IST timezone
ist = pytz.timezone('Asia/Kolkata')

# Prepare a list to store file info rows
data = []

# Loop through each file in the previously fetched file_list
for i, file in enumerate(file_list, 1):
    name = file['title']  # File name
    owner = file.get('owners')[0]['displayName'] if file.get('owners') else 'N/A'  # Owner name

    # Get modified date string in UTC (ISO format)
    modified_date_utc = file['modifiedDate']

    # Convert the UTC datetime string to a timezone-aware pandas datetime object
    dt_utc = pd.to_datetime(modified_date_utc).tz_convert('UTC')

    # Convert UTC time to IST timezone
    dt_ist = dt_utc.tz_convert(ist)

    # Format datetime in dd/mm/yyyy hh:mm:ss AM/PM IST format
    modified_date_ist_str = dt_ist.strftime('%d/%m/%Y %I:%M:%S %p IST')

    # Get file size in bytes (if available)
    size_bytes = int(file['fileSize']) if 'fileSize' in file else 0

    # Convert file size to human-readable string
    size_hr = human_readable_size(size_bytes)

    # Append all collected info as a row (with serial number)
    data.append([i, name, owner, modified_date_ist_str, size_hr])

# Create a pandas DataFrame with the collected data and column headers
df = pd.DataFrame(data, columns=['S.No', 'Name', 'Owner', 'Last Modified (IST)', 'File Size'])

# Display the DataFrame as a table
df
```

## Top Interview Questions and Answers:

Here are some key interview questions based on the concepts in your notebook:

*   **Q1: What is Linear Regression and when would you use it?**
    *   **A1:** Linear Regression is a supervised learning algorithm used to predict a continuous target variable based on one or more input features. You would use it when you believe there is a linear relationship between the features and the target.
*   **Q2: Explain the difference between Simple and Multiple Linear Regression.**
    *   **A2:** Simple Linear Regression uses one independent variable, while Multiple Linear Regression uses two or more independent variables to predict the dependent variable.
*   **Q3: What is the purpose of the `.fit()` method in scikit-learn's Linear Regression?**
    *   **A3:** The `.fit()` method is used to train the model. It learns the relationship between the input features and the target variable from the provided data.
*   **Q4: What is pickling in Python and why is it useful in machine learning?**
    *   **A4:** Pickling is the process of serializing and deserializing Python objects. In machine learning, it's useful for saving a trained model to a file so you can load and use it later without needing to retrain it from scratch.
*   **Q5: How do you make a prediction with a trained Linear Regression model in scikit-learn?**
    *   **A5:** You use the `.predict()` method, passing the input feature(s) you want to predict for. The input should be in the same format (e.g., a 2D array) as the data used for training.
*   **Q6: What is the significance of converting a pandas DataFrame to a numpy array before training a model?**
    *   **A6:** While scikit-learn often handles this automatically, converting to a numpy array can sometimes improve performance and ensures the data is in a format that many machine learning libraries are optimized for. It also discards potentially unnecessary information like column names and index.
*   **Q7: How would you handle categorical features if you were to use them in a Linear Regression model?**
    *   **A7:** Linear Regression requires numerical input. Categorical features need to be encoded into numerical representations, commonly using techniques like One-Hot Encoding.
*   **Q8: What are some common metrics used to evaluate a Linear Regression model?**
    *   **A8:** Common metrics include Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared.

    *   **A9:** Splitting data allows you to evaluate how well your model generalizes to unseen data. You train the model on the training set and test its performance on the testing set to get an unbiased estimate of its accuracy.
