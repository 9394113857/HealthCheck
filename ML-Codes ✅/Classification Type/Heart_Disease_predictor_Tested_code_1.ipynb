```python
# Load the trained Logistic Regression model from the saved file
model = pickle.load(open('heart_model.pkl', 'rb'))

# Set placeholder values
placeholder_age = 45
placeholder_sex = 1  # 0 for female, 1 for male
placeholder_cp = 2
placeholder_trestbps = 130
placeholder_chol = 220
placeholder_fbs = 0
placeholder_restecg = 1
placeholder_thalach = 160
placeholder_exang = 0
placeholder_oldpeak = 1.5
placeholder_slope = 2
placeholder_ca = 0
placeholder_thal = 2

# Take all inputs from the user
age = int(input(f"Enter age (or press Enter to use placeholder {placeholder_age}): ") or placeholder_age)
sex = int(input(f"Enter sex (0 for female, 1 for male, or press Enter to use placeholder {placeholder_sex}): ") or placeholder_sex)
cp = int(input(f"Enter chest pain type (0-3, or press Enter to use placeholder {placeholder_cp}): ") or placeholder_cp)
trestbps = int(input(f"Enter resting blood pressure (or press Enter to use placeholder {placeholder_trestbps}): ") or placeholder_trestbps)
chol = int(input(f"Enter serum cholesterol level (or press Enter to use placeholder {placeholder_chol}): ") or placeholder_chol)
fbs = int(input(f"Enter fasting blood sugar (0 or 1, or press Enter to use placeholder {placeholder_fbs}): ") or placeholder_fbs)
restecg = int(input(f"Enter resting electrocardiographic results (0-2, or press Enter to use placeholder {placeholder_restecg}): ") or placeholder_restecg)
thalach = int(input(f"Enter maximum heart rate achieved (or press Enter to use placeholder {placeholder_thalach}): ") or placeholder_thalach)
exang = int(input(f"Enter exercise-induced angina (0 or 1, or press Enter to use placeholder {placeholder_exang}): ") or placeholder_exang)
oldpeak = float(input(f"Enter ST depression induced by exercise relative to rest (or press Enter to use placeholder {placeholder_oldpeak}): ") or placeholder_oldpeak)
slope = int(input(f"Enter the slope of the peak exercise ST segment (0-2, or press Enter to use placeholder {placeholder_slope}): ") or placeholder_slope)
ca = int(input(f"Enter number of major vessels (0-3, or press Enter to use placeholder {placeholder_ca}): ") or placeholder_ca)
thal = int(input(f"Enter thalassemia (1-3, or press Enter to use placeholder {placeholder_thal}): ") or placeholder_thal)

# Create a dictionary with the user input
user_input = {
    'age': [age],
    'sex': [sex],
    'cp': [cp],
    'trestbps': [trestbps],
    'chol': [chol],
    'fbs': [fbs],
    'restecg': [restecg],
    'thalach': [thalach],
    'exang': [exang],
    'oldpeak': [oldpeak],
    'slope': [slope],
    'ca': [ca],
    'thal': [thal]
}

# Convert the user input to a DataFrame
user_input_df = pd.DataFrame.from_dict(user_input)

# Reshape the input to be a 2D array
reshaped_input = user_input_df.values.reshape(1, -1)

# Make predictions using the loaded model
prediction = model.predict(reshaped_input)

# Create a table with the user information
table = [
    ['S.No', 'Attribute', 'Value'],
    [1, 'Age', age],
    [2, 'Sex', 'Male' if sex == 1 else 'Female'],
    [3, 'Chest Pain Type', cp],
    [4, 'Resting Blood Pressure', trestbps],
    [5, 'Serum Cholesterol Level', chol],
    [6, 'Fasting Blood Sugar', fbs],
    [7, 'Resting Electrocardiographic Results', restecg],
    [8, 'Maximum Heart Rate Achieved', thalach],
    [9, 'Exercise-Induced Angina', exang],
    [10, 'ST Depression Induced by Exercise Relative to Rest', oldpeak],
    [11, 'Slope of the Peak Exercise ST Segment', slope],
    [12, 'Number of Major Vessels Colored by Fluoroscopy', ca],
    [13, 'Thalassemia', thal]
]

# Print the table
print(tabulate(table, headers='firstrow', tablefmt='fancy_grid'))

# Print the predicted result with formatting
print()
if prediction[0] == 0:

    print(f"\033[92m\033[1mThe user, aged {age}, is not likely to have heart disease.\033[0m")
else:
    print(f"\033[91m\033[1mThe user, aged {age}, is likely to have heart disease.\033[0m")
```

```python
# Function to print a formatted table
def print_table(header, data):
    # Calculate column widths
    col_widths = [max(len(str(row[i])) for row in data) for i in range(len(header))]

    # Print the header
    print("┌" + "─" * (sum(col_widths) + len(col_widths) * 3 - 1) + "┐")
    print("│", end="")
    for i, col_name in enumerate(header):
        print(f" {col_name.ljust(col_widths[i])} │", end="")
    print()

    # Print separator
    print("├" + "─" * (sum(col_widths) + len(col_widths) * 3 - 1) + "┤")

    # Print data rows
    for row in data:
        print("│", end="")
        for i, value in enumerate(row):
            print(f" {str(value).ljust(col_widths[i])} │", end="")
        print()

    # Print bottom border
    print("└" + "─" * (sum(col_widths) + len(col_widths) * 3 - 1) + "┘")

# Define header and data
header = ["S.No", "Attribute", "Range (No)", "Range (Yes)"]
data = [
    [1, "Age", "45", "60"],
    [2, "Sex", "Male", "Male"],
    [3, "Chest Pain Type", "2", "3"],
    [4, "Resting Blood Pressure", "130", "140"],
    [5, "Serum Cholesterol Level", "220", "260"],
    [6, "Fasting Blood Sugar", "0", "1"],
    [7, "Resting Electrocardiographic Results", "1", "2"],
    [8, "Maximum Heart Rate Achieved", "160", "120"],
    [9, "Exercise-Induced Angina", "0", "1"],
    [10, "ST Depression Induced by Exercise Relative to Rest", "1.5", "3"],
    [11, "Slope of the Peak Exercise ST Segment", "2", "2"],
    [12, "Number of Major Vessels Colored by Fluoroscopy", "0", "1"],
    [13, "Thalassemia", "2", "3"]
]

# Print the table
print_table(header, data)
```
