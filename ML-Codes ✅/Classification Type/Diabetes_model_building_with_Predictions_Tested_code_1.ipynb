```python
# Step 6: Extract file metadata and display as a table with Last Modified time converted to IST
# and formatted as dd/mm/yyyy hh:mm:ss AM/PM IST

import pandas as pd
import pytz  # For timezone conversions

# Helper function to convert bytes to human-readable format
def human_readable_size(size):
    for unit in ['B', 'KB', 'MB', 'GB']:
        if size < 1024:
            return f"{size:.1f} {unit}"
        size /= 1024
    return f"{size:.1f} TB"

# Define IST timezone
ist = pytz.timezone('Asia/Kolkata')

# Prepare a list to store file info rows
data = []

# Loop through each file in the previously fetched file_list
for i, file in enumerate(file_list, 1):
    name = file['title']  # File name
    owner = file.get('owners')[0]['displayName'] if file.get('owners') else 'N/A'  # Owner name

    # Get modified date string in UTC (ISO format)
    modified_date_utc = file['modifiedDate']

    # Convert the UTC datetime string to a timezone-aware pandas datetime object
    dt_utc = pd.to_datetime(modified_date_utc).tz_convert('UTC')

    # Convert UTC time to IST timezone
    dt_ist = dt_utc.tz_convert(ist)

    # Format datetime in dd/mm/yyyy hh:mm:ss AM/PM IST format
    modified_date_ist_str = dt_ist.strftime('%d/%m/%Y %I:%M:%S %p IST')

    # Get file size in bytes (if available)
    size_bytes = int(file['fileSize']) if 'fileSize' in file else 0

    # Convert file size to human-readable string
    size_hr = human_readable_size(size_bytes)

    # Append all collected info as a row (with serial number)
    data.append([i, name, owner, modified_date_ist_str, size_hr])

# Create a pandas DataFrame with the collected data and column headers
df = pd.DataFrame(data, columns=['S.No', 'Name', 'Owner', 'Last Modified (IST)', 'File Size'])

# Display the DataFrame as a table
df
```
