This code will predict whether the given emails are spam or not. It uses the trained clf model and displays the predictions accordingly.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/9394113857/Data-Sets/9882400f949ec59791995b8be588891f9e0daf48/spam.csv', delimiter=',', encoding='utf-8')

# Add a 'Spam' column
data['Spam'] = data['Category'].apply(lambda x: 1 if x == 'spam' else 0)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.Message, data.Spam, test_size=0.25)

# Define the pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])

# Train the model
clf.fit(X_train, y_train)

# Function to predict spam or not spam based on user input
def predict_spam():
    print("Enter an email message:")
    email = input("Email: ").strip()
    default_response = 'Sounds great! Are you home now?'  # Default placeholder

    if not email or email.lower() == 'yes':
        email = default_response

    prediction = clf.predict([email])
    result = "Spam" if prediction[0] == 1 else "Not Spam"
    print("Prediction:", result)

# Test the function
predict_spam()
```

```python
import joblib

# Save the trained model
joblib.dump(clf, 'spam_detection_model.pkl')

# Load the saved model
loaded_model = joblib.load('spam_detection_model.pkl')

# Function to predict spam or not spam based on user input using the loaded model
def predict_spam_loaded_model():
    print("Enter an email message:")
    email = input("Email: ").strip()
    default_response = 'Sounds great! Are you home now?'  # Default placeholder

    if not email or email.lower() == 'yes':
        email = default_response

    prediction = loaded_model.predict([email])
    result = "Spam" if prediction[0] == 1 else "Not Spam"
    print("Prediction:", result)

# Test the function with the loaded model
predict_spam_loaded_model()
```

```python
import joblib
import os
from rich.console import Console
from rich.table import Table
from rich import box  # Importing box styles
from datetime import datetime
import pytz
import warnings

# Suppress any warnings
warnings.filterwarnings("ignore", category=UserWarning)

# List all files in the current directory and filter out the model files
def list_model_files():
    files = [f for f in os.listdir() if f.endswith('.pkl')]
    return files

# Automatically detect available model file(s)
model_files = list_model_files()

# If no model file is found, exit
if not model_files:
    print("Error: No model files found in the current directory.")
    exit()

# If only one model file is found, use it automatically
if len(model_files) == 1:
    default_model_name = model_files[0]
    print(f"Using default model file: {default_model_name}")

# If multiple model files are found, prompt the user to select one
else:
    print("Multiple model files found:")
    for i, model in enumerate(model_files, start=1):
        print(f"{i}. {model}")

    # Ask the user to select a model
    try:
        choice = int(input(f"\nEnter the number corresponding to the model you want to use (1-{len(model_files)}): "))
        if 1 <= choice <= len(model_files):
            default_model_name = model_files[choice - 1]
            print(f"Using selected model file: {default_model_name}")
        else:
            print("Invalid choice. Exiting.")
            exit()
    except ValueError:
        print("Invalid input. Exiting.")
        exit()

# Try to load the model file
try:
    loaded_model = joblib.load(default_model_name)
except Exception as e:
    print(f"Error loading the model: {e}")
    exit()  # Exit if the model couldn't be loaded

# Function to get current time in IST and formatted date/time string
def get_current_time_ist():
    ist = pytz.timezone('Asia/Kolkata')  # Set IST timezone
    now_ist = datetime.now(ist)  # Get the current time in IST
    formatted_date_time = now_ist.strftime('%Y-%m-%d %H:%M:%S')  # Format the date and time
    return formatted_date_time

# Function to predict spam or not spam based on user input using the loaded model
def predict_spam_loaded_model():
    console = Console()

    # Get current time for displaying
    current_time = get_current_time_ist()

    # Direct and simple email message input prompt with extra line breaks for clarity
    console.print("\n[bold magenta on white] Enter an Email Message to Check Spam Status: [/bold magenta on white]")
    email = input().strip()  # Get user input directly

    # Default response in case of empty input
    default_response = 'Sounds great! Are you home now?'

    # If the input is empty or contains only 'yes', use the default message
    if not email or email.lower() == 'yes':
        email = default_response

    # Add spacing before displaying the user input
    console.print(f"\n[bold cyan]User Input:[/bold cyan] [italic yellow]{email}[/italic yellow]", style="bold")
    console.print("\n" + "-" * 60)  # Separator for better visual structure

    # Display current time of execution in the output
    console.print(f"[bold green]Execution Time: {current_time}[/bold green]")
    console.print("\n" + "-" * 60)  # Separator for better visual structure

    # Make a prediction using the loaded model
    prediction = loaded_model.predict([email])

    # Interpret the result
    result = "Spam" if prediction[0] == 1 else "Not Spam"

    # Add some space before showing prediction result
    console.print(f"\n[bold blue on white]Prediction Result:[/bold blue on white] [bold red]{result}[/bold red]\n")
    console.print("\n" + "-" * 60)  # Separator for better visual structure

    # Create and style the result table with more separation and better alignment
    table = Table(title="[bold green]Spam Detection Result[/bold green]", box=box.ROUNDED, show_header=True)

    # Adding styled columns with clear text alignment and space between them
    table.add_column("Email Message", style="bold white", justify="center", width=40)
    table.add_column("Prediction", style="bold yellow", justify="center", width=20)

    # Add the email message and prediction to the table
    table.add_row(email, result)

    # Print the result in a well-formatted table with extra spacing
    console.print("\n" + "-" * 60)  # Final separator
    console.print(table)

# Test the function with the loaded model
predict_spam_loaded_model()
```

Here are some potential interview questions and answers based on the concepts covered in this notebook:

**Q1: What is the main task being performed in this notebook?**
A1: The main task is building a spam email detection model using machine learning.

**Q2: What type of machine learning problem is spam detection considered?**
A2: Spam detection is a supervised learning problem, specifically binary classification, where the goal is to classify emails into one of two categories: "spam" or "not spam".

**Q3: What dataset is being used for this task?**
A3: The notebook uses a dataset containing email messages and their corresponding categories ('spam' or 'ham'), loaded from a CSV file hosted on GitHub.

**Q4: What is the purpose of the 'Spam' column created in the DataFrame?**
A4: The 'Spam' column is a numerical representation of the 'Category' column, where 'spam' is mapped to 1 and 'ham' (not spam) is mapped to 0. This is done to prepare the data for the machine learning model, which typically works with numerical inputs.

**Q5: How is the data split for training and testing the model?**
A5: The data is split into training and testing sets using `train_test_split` from scikit-learn. 25% of the data is allocated for testing (`test_size=0.25`).

**Q6: What technique is used to convert the text data (email messages) into a format suitable for the machine learning model?**
A6: The `CountVectorizer` from scikit-learn is used to convert the text data into a matrix of token counts. This represents each email as a numerical vector based on the frequency of words.

**Q7: Which machine learning algorithm is used for classification in this notebook? Why was this specific type of Naive Bayes classifier chosen?**
A7: A Multinomial Naive Bayes classifier (`MultinomialNB`) is used. This type is suitable for discrete data like word counts, which are generated by the `CountVectorizer`.

**Q8: Explain the role of the `Pipeline` in this notebook.**
A8: The `Pipeline` is used to chain multiple steps together, specifically the `CountVectorizer` (text vectorization) and the `MultinomialNB` classifier (model training). This simplifies the workflow and ensures that the same preprocessing steps are applied consistently to both the training and testing data.

**Q9: How is the trained model saved and loaded?**
A9: The `joblib` library is used to save the trained model to a file (`spam_detection_model.pkl`) and load it back later. This allows for reusing the trained model without retraining.

A10: The notebook uses the `score()` method of the trained pipeline (`clf.score(X_test, y_test)`) to evaluate the model's accuracy on the test set.
